## Get good clean data to work with
Sort by date: StormEvents_locations-ftp<br />
List only files with data: Some filenames have no data<br />
Create a list of the file: called LIST<br />    

```python
import os 
import os.path
LOCATIONS = []
PATH = "/notebooks/examples/pub/data/swdi/stormevents/csvfiles/"
files = os.listdir(PATH)
files = sorted(files)
for file in files:
    if "locations" in file:
        SZ = os.path.getsize(PATH+file)
        if SZ>300:
            print (file,":",SZ)
            LOCATIONS.append(PATH+file)
        if SZ<300:
            # Remove and files with a size <300
            # This removes empty files that contain only a header
            os.remove(PATH+file)
```
